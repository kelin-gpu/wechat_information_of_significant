```python
import pandas as pd
import re
import datetime
```

```python
# 1️⃣ 读取原始 CSV
df = pd.read_csv("ama_notdo.csv")
df
```

```python
# 2️⃣ 新增人类可读时间列
def convert_time(ts):
    try:
        # 判断是秒还是毫秒时间戳
        ts = float(ts)
        if ts > 1e12:  # 毫秒级
            ts /= 1000
        return datetime.datetime.fromtimestamp(ts)
    except:
        return None

df["readable_time"] = df["create_time"].apply(convert_time)
```

```python
# 3️⃣ 定义函数：清理消息并拆分 sender + content
def parse_message(msg):
    if not isinstance(msg, str) or len(msg.strip()) == 0:
        return (None, None)

    # 去除二进制内容
    if msg.startswith("b'") or msg.startswith('b"') or re.search(r'\\x[0-9a-fA-F]{2}', msg):
        return (None, None)

    msg = msg.strip()
    msg = re.sub(r'[\x00-\x1f\x7f]+', '', msg)  # 去除控制符

    # 尝试匹配格式：wxid_xxx:\n内容
    match = re.match(r'([\w\-]+):\s*\n?(.*)', msg)
    if match:
        sender = match.group(1).strip()
        content = match.group(2).strip()
    else:
        sender = None
        content = msg.strip()

    # 去除多余换行
    content = re.sub(r'\s+', ' ', content)

    # 若消息为空则返回 None
    if not content:
        return (sender, None)

    return (sender, content)

# 4️⃣ 应用拆分
df[["User", "text_content"]] = df["message_content"].apply(
    lambda x: pd.Series(parse_message(x))
)
df
```

```python
# 5️⃣ 仅保留需要的列
clean_df = df[["create_time", "readable_time", "User", "text_content"]].dropna(subset=["text_content"])
clean_df = clean_df.dropna(subset=["User", "text_content"])
clean_df
```

```python
# 6️⃣ 保存结果
clean_df.to_csv("ama_message_done.csv", index=False, encoding="utf-8-sig")

print(f"✅ 已成功提取 {len(clean_df)} 条有效消息。")
print("🎉 输出文件: ama_done.csv")
print(clean_df.head(10))
```
