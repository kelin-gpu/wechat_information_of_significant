```python
import joblib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 1. 加载模型和编码器
model = joblib.load('acc_model.pkl')
encoder = joblib.load('encoder.pkl')
```

```python
# 2. 读取数据
df = pd.read_csv('ama_all_norm.csv')
```

```python
# 3. 做相同的预处理
X_new = encoder.transform(df.iloc[0:100, 0:3])

# 4. 使用模型进行预测
predictions = model.predict(X_new)
actual_values = df.iloc[0:100, 3].values
print(predictions,"\n",actual_values)
```
